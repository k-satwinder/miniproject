```python
item = "19M"
item[-1]
def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M",""))
        res = res * 1024
        return res
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k",""))
        res = res * 1024
        return res
    else:
        return str(np.nan)
```

```python
def clean_android_version(val):
    if pd.isna(val) or str(val).strip() == "":
        return None
    
    val = str(val).strip()
    
    # Handle ranges like "4.0.3 - 7.1.1" → keep higher value
    if "-" in val:
        val = val.split("-")[-1].strip()
    
    # Remove non-numeric characters (like 'W')
    val = ''.join(ch for ch in val if (ch.isdigit() or ch == '.'))
    
    if val == "":
        return None
    
    try:
        num = float(val)
        return round(num, 1)   # keep only one decimal place
    except:
        return None
```

```python
df_copy['Android Ver'] = df_copy['Android Ver'].apply(clean_android_version)
```
